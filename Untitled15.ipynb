```python
import pandas as pd
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import accuracy_score, recall_score, classification_report, confusion_matrix


# Features and labels
X = df.drop(columns=['participant', 'participant_type', 'session', 'emotion', 'label'])
y = df['label']  # 1 for MDD, 0 for HC

# Number of folds
k = 5  # Adjust based on your dataset size

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

# Initialize lists to store performance metrics
accuracy_list = []
sensitivity_list = []  # Recall for positive class (MDD)
specificity_list = []  # Recall for negative class (HC)

# Function to select sensitive and specific features
def select_sensitive_specific_features(X_train, y_train, num_features=5):
    # Perform ANOVA F-test
    selector = SelectKBest(score_func=f_classif, k=num_features)
    selector.fit(X_train, y_train)
    selected_features = X_train.columns[selector.get_support()]
    return list(selected_features)

# Loop over each fold
for fold, (train_index, test_index) in enumerate(skf.split(X, y), 1):
    print(f"\nFold {fold}")

    # Split data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Identify sensitive features (features that best distinguish MDD from HC)
    sensitive_features = select_sensitive_specific_features(X_train, y_train, num_features=5)

    # Identify specific features (features that best distinguish HC from MDD)
    # For specific features, invert the labels (1 becomes 0 and vice versa)
    specific_features = select_sensitive_specific_features(X_train, 1 - y_train, num_features=5)

    # Root Node: SVM with sensitive features
    root_svm = SVC(kernel='linear', probability=True, random_state=42)
    root_svm.fit(X_train[sensitive_features], y_train)

    # Predict on test set
    root_predictions = root_svm.predict(X_test[sensitive_features])
    root_probabilities = root_svm.predict_proba(X_test[sensitive_features])[:, 1]

    # Left Child Node: SVM with specific features
    # Train left_svm on the entire training data using specific features
    left_svm = SVC(kernel='linear', probability=True, random_state=42)
    left_svm.fit(X_train[specific_features], y_train)

    # For samples predicted as 'normal' by the root node, use left_svm to refine the prediction
    normal_indices = np.where(root_predictions == 0)[0]
    if len(normal_indices) > 0:
        left_predictions = left_svm.predict(X_test.iloc[normal_indices][specific_features])
        # Update predictions
        root_predictions[normal_indices] = left_predictions
    else:
        print("No samples predicted as 'normal' by root SVM in this fold.")

    # Evaluate the model
    accuracy = accuracy_score(y_test, root_predictions)
    sensitivity = recall_score(y_test, root_predictions, pos_label=1)
    specificity = recall_score(y_test, root_predictions, pos_label=0)

    accuracy_list.append(accuracy)
    sensitivity_list.append(sensitivity)
    specificity_list.append(specificity)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Sensitivity (Recall for MDD): {sensitivity:.2f}")
    print(f"Specificity (Recall for HC): {specificity:.2f}")

    # Optional: Display classification report for each fold
    print("\nClassification Report:")
    print(classification_report(y_test, root_predictions, target_names=['HC', 'MDD']))

# After cross-validation loop
print("\nOverall Performance:")
print(f"Average Accuracy: {np.mean(accuracy_list):.2f} ± {np.std(accuracy_list):.2f}")
print(f"Average Sensitivity (MDD Recall): {np.mean(sensitivity_list):.2f} ± {np.std(sensitivity_list):.2f}")
print(f"Average Specificity (HC Recall): {np.mean(specificity_list):.2f} ± {np.std(specificity_list):.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns

# Features and labels
X = df.drop(columns=['participant', 'participant_type', 'session', 'emotion', 'label'])
y = df['emotion']  # 'Positive', 'Neutral', 'Negative'

# Encode labels if they are not numeric
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_encoded = le.fit_transform(y)  # 0: Negative, 1: Neutral, 2: Positive

# Number of folds
k = 5

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=k, shuffle=True, random_state=42)

# Initialize lists to store performance metrics
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

# Initialize a DataFrame to store feature importances
feature_importances = pd.DataFrame()

# Feature selection: Select top N features
def select_features(X_train, y_train, num_features=10):
    selector = SelectKBest(score_func=f_classif, k=num_features)
    selector.fit(X_train, y_train)
    selected_features = X_train.columns[selector.get_support()]
    return list(selected_features)

# Hyperparameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [None, 10, 50],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'class_weight': ['balanced']  # To handle class imbalance
}

for fold, (train_index, test_index) in enumerate(skf.split(X, y_encoded), 1):
    print(f"\nFold {fold}")

    # Split data
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y_encoded[train_index], y_encoded[test_index]

    # Feature Selection
    selected_features = select_features(X_train, y_train, num_features=10)
    print(f"Selected Features: {selected_features}")

    # Hyperparameter Tuning with GridSearchCV
    rf = RandomForestClassifier(random_state=42)
    grid_search = GridSearchCV(
        estimator=rf,
        param_grid=param_grid,
        cv=3,
        scoring='accuracy',
        n_jobs=-1,
        verbose=0
    )
    grid_search.fit(X_train[selected_features], y_train)
    best_rf = grid_search.best_estimator_
    print(f"Best Parameters: {grid_search.best_params_}")

    # Train the best model
    best_rf.fit(X_train[selected_features], y_train)

    # Predict on test set
    y_pred = best_rf.predict(X_test[selected_features])

    # Evaluate the model
    accuracy = accuracy_score(y_test, y_pred)
    precision = recall_score(y_test, y_pred, average='macro')  # Macro average for multi-class
    recall = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')

    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision (Macro): {precision:.2f}")
    print(f"Recall (Macro): {recall:.2f}")
    print(f"F1-Score (Macro): {f1:.2f}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=le.classes_))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - Fold {fold}')
    plt.show()

    # Feature Importances
    fold_importances = pd.DataFrame({
        'feature': selected_features,
        'importance': best_rf.feature_importances_,
        'fold': fold
    })
    feature_importances = pd.concat([feature_importances, fold_importances], ignore_index=True)

# Overall Performance
print("\nOverall Performance:")
print(f"Average Accuracy: {np.mean(accuracy_list):.2f} ± {np.std(accuracy_list):.2f}")
print(f"Average Precision (Macro): {np.mean(precision_list):.2f} ± {np.std(precision_list):.2f}")
print(f"Average Recall (Macro): {np.mean(recall_list):.2f} ± {np.std(recall_list):.2f}")
print(f"Average F1-Score (Macro): {np.mean(f1_list):.2f} ± {np.std(f1_list):.2f}")

# Aggregating Feature Importances
average_importances = feature_importances.groupby('feature')['importance'].mean().sort_values(ascending=False)
print("\nTop Features Based on Average Importance:")
print(average_importances.head(10))

# Plotting Feature Importances
plt.figure(figsize=(10,6))
sns.barplot(x=average_importances.values[:10], y=average_importances.index[:10], palette='viridis')
plt.title('Top 10 Feature Importances Across Folds')
plt.xlabel('Average Importance')
plt.ylabel('Feature')
plt.show()
```

```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.svm import SVC
from sklearn.metrics import (
    classification_report,
    confusion_matrix,
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Suppress warnings for cleaner output
import warnings
warnings.filterwarnings('ignore')


# Separate Features and Labels
X = df.drop(columns=['participant', 'emotion', 'session', 'label','participant_type'])  # Drop 'emotion', classify 'participant_type'
y = df['label']  # Classify between MDD and HC


# Verify Label Encoding
print("\nLabel Classes:")
print(le.classes_)
print("\nEncoded Labels (first 10):")
print(y_encoded[:10])

# -----------------------------
# 2. Feature Selection Function
# -----------------------------

def select_features(X_train, y_train, k=10):
    """
    Select the top k features based on ANOVA F-test.

    Parameters:
    - X_train: Training features
    - y_train: Training labels
    - k: Number of top features to select

    Returns:
    - selected_features: List of selected feature names
    """
    selector = SelectKBest(score_func=f_classif, k=k)
    selector.fit(X_train, y_train)
    selected_features = X_train.columns[selector.get_support()]
    return selected_features.tolist()

# -----------------------------
# 3. Cross-Validation and SVM Implementation
# -----------------------------

# Define the number of folds
k_folds = 5

# Initialize StratifiedKFold
skf = StratifiedKFold(n_splits=k_folds, shuffle=True, random_state=42)

# Initialize lists to store performance metrics
accuracy_list = []
precision_list = []
recall_list = []
f1_list = []

# Initialize a DataFrame to store feature importances
feature_importances = pd.DataFrame()

# Create a Pipeline with Scaling and SVM
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('svm', SVC(probability=True, decision_function_shape='ovr', random_state=42))
])

# Define hyperparameter grid for GridSearchCV
param_grid = {
    'svm__C': [0.1, 1, 10, 100],
    'svm__kernel': ['linear'],  # Using 'linear' for feature importances
    'svm__class_weight': ['balanced'],  # To handle class imbalance
}

# Loop over each fold
for fold, (train_index, test_index) in enumerate(skf.split(X, y_encoded), 1):
    print(f"\n--- Fold {fold} ---")

    # Split data into training and testing sets
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y_encoded[train_index], y_encoded[test_index]

    # Feature Selection
    selected_features = select_features(X_train, y_train, k=10)
    print(f"Selected Features: {selected_features}")

    # Initialize GridSearchCV with the pipeline
    grid_search = GridSearchCV(
        estimator=pipeline,
        param_grid=param_grid,
        cv=3,
        scoring='accuracy',
        n_jobs=-1,
        verbose=0
    )

    # Fit GridSearchCV
    try:
        grid_search.fit(X_train[selected_features], y_train)
    except ValueError as e:
        print(f"Error during GridSearchCV fitting: {e}")
        print("Attempting to debug by setting error_score='raise'")
        # Re-initialize GridSearchCV with error_score='raise' to see the actual error
        grid_search = GridSearchCV(
            estimator=pipeline,
            param_grid=param_grid,
            cv=3,
            scoring='accuracy',
            n_jobs=-1,
            verbose=0,
            error_score='raise'
        )
        grid_search.fit(X_train[selected_features], y_train)

    best_pipeline = grid_search.best_estimator_
    print(f"Best Parameters: {grid_search.best_params_}")

    # Predict on the test set
    y_pred = best_pipeline.predict(X_test[selected_features])

    # Calculate Performance Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='macro', zero_division=0)
    recall = recall_score(y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='macro', zero_division=0)

    # Append metrics to lists
    accuracy_list.append(accuracy)
    precision_list.append(precision)
    recall_list.append(recall)
    f1_list.append(f1)

    # Display Metrics
    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision (Macro): {precision:.2f}")
    print(f"Recall (Macro): {recall:.2f}")
    print(f"F1-Score (Macro): {f1:.2f}")

    # Classification Report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=le.classes_))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(f'Confusion Matrix - Fold {fold}')
    plt.show()

    # Feature Importances (Only for Linear Kernel)
    svm_best = best_pipeline.named_steps['svm']
    if svm_best.kernel == 'linear':
        coef = svm_best.coef_
        # For multi-class, coef_ is a matrix (n_classes, n_features)
        for class_idx, class_label in enumerate(le.classes_):
            fold_importances = pd.DataFrame({
                'feature': selected_features,
                'importance': np.abs(coef[class_idx]),
                'class': class_label,
                'fold': fold
            })
            feature_importances = pd.concat([feature_importances, fold_importances], ignore_index=True)
    else:
        print("Feature importances not available for non-linear kernels.")

# -----------------------------
# 4. Overall Performance Metrics
# -----------------------------

print("\n=== Overall Performance ===")
print(f"Average Accuracy: {np.mean(accuracy_list):.2f} ± {np.std(accuracy_list):.2f}")
print(f"Average Precision (Macro): {np.mean(precision_list):.2f} ± {np.std(precision_list):.2f}")
print(f"Average Recall (Macro): {np.mean(recall_list):.2f} ± {np.std(recall_list):.2f}")
print(f"Average F1-Score (Macro): {np.mean(f1_list):.2f} ± {np.std(f1_list):.2f}")

# -----------------------------
# 5. Feature Importance Analysis
# -----------------------------

if not feature_importances.empty:
    # Calculate average importance per feature per class
    average_importances = feature_importances.groupby(['feature', 'class'])['importance'].mean().reset_index()

    # Pivot the DataFrame for better visualization
    pivot_importances = average_importances.pivot(index='feature', columns='class', values='importance').fillna(0)

    # Sort features based on overall importance
    pivot_importances['Overall'] = pivot_importances.sum(axis=1)
    pivot_importances = pivot_importances.sort_values(by='Overall', ascending=False)

    # Display Top 10 Features
    top_features = pivot_importances.head(10)
    print("\nTop 10 Features Based on Average Importance:")
    print(top_features)

    # Plot Feature Importances
    top_features_sorted = top_features.sort_values(by='Overall', ascending=True)
    top_features_sorted.drop(columns='Overall').plot(kind='barh', figsize=(10,8), stacked=True, colormap='viridis')
    plt.title('Top 10 Feature Importances Across Classes')
    plt.xlabel('Average Importance')
    plt.ylabel('Feature')
    plt.legend(title='Emotion', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("No feature importances available. Ensure that SVMs with linear kernels were used.")

# -----------------------------
# 6. Optional: Statistical Testing for Feature Associations
# -----------------------------

from scipy.stats import f_oneway

def perform_anova(df, feature, label_col='emotion'):
    """
    Perform ANOVA F-test for a given feature across different groups.

    Parameters:
    - df: DataFrame containing the data
    - feature: Feature name to test
    - label_col: Column name for labels

    Returns:
    - F-statistic and p-value
    """
    groups = [df[df[label_col] == emotion][feature] for emotion in le.classes_]
    F, p = f_oneway(*groups)
    return F, p

# List of features to test
features_to_test = selected_features  # Or specify a list of interest

anova_results = {}
for feature in features_to_test:
    F, p = perform_anova(df, feature)
    anova_results[feature] = {'F-statistic': F, 'p-value': p}

# Convert to DataFrame
anova_df = pd.DataFrame(anova_results).T
print("\nANOVA Results for Selected Features:")
print(anova_df)

# Identify significant features (p < 0.05)
significant_features = anova_df[anova_df['p-value'] < 0.05].index.tolist()
print("\nSignificant Features (p < 0.05):", significant_features)

# Plot Boxplots for Significant Features
if significant_features:
    for feature in significant_features:
        plt.figure(figsize=(8,6))
        sns.boxplot(x='emotion', y=feature, data=df)
        plt.title(f'Boxplot of {feature} Across Emotions')
        plt.xlabel('Emotion')
        plt.ylabel(feature)
        plt.show()
else:
    print("No significant features found based on ANOVA.")

# -----------------------------
# 7. Summary and Recommendations
# -----------------------------

print("\n=== Summary ===")
print(f"Total Folds: {k_folds}")
print(f"Average Accuracy: {np.mean(accuracy_list):.2f} ± {np.std(accuracy_list):.2f}")
print(f"Average Precision (Macro): {np.mean(precision_list):.2f} ± {np.std(precision_list):.2f}")
print(f"Average Recall (Macro): {np.mean(recall_list):.2f} ± {np.std(recall_list):.2f}")
print(f"Average F1-Score (Macro): {np.mean(f1_list):.2f} ± {np.std(f1_list):.2f}")


```
